```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-10, 10, 400)  # 400 puntos entre -10 y 10
y = x**2  # Función cuadrática

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = x^2$', color='b')
plt.axhline(0, color='black', linewidth=1)  # Eje X
plt.axvline(0, color='black', linewidth=1)  # Eje Y
plt.grid(True, linestyle='--', alpha=0.6)  # Cuadrícula
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de y = x^2")
plt.legend()
plt.show()
```
