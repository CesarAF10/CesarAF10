```python
from tabulate import tabulate
import matplotlib.pyplot as plt
a=17
c=21
m=2**10
x0=123
def generador(a, c, m, x0):
    x=(a*x0+c)%m
    return x
lista=[]
muestra=[]
for i in range(10000):
  nuevo=generador(a, c, m, x0)
  lista.append(nuevo/m)
  muestra.append([i, x0, nuevo/m])
  x0=nuevo
plt.plot(lista)
plt.hist(lista)
plt.show()
encabezador=["i","x_i", "x_{i+1}"]
print(tabulate(muestra, headers=encabezador))
```

Generador congruencial para obtener una muestra de 10000 numeros aleatorios, obtener media, varianza y graficar
